```python
from binance.client import Client
import pandas as pd
```

```python
def get_actual_data(
    symbol: str,
    time_frame: str = "5m",
    begin_date: str = "2023-10-01T00:00:00Z",
    end_date: str = "10 hours ago UTC",
) -> pd.DataFrame:
    client = Client()
    df = pd.DataFrame(client.get_historical_klines(symbol, time_frame, begin_date, end_date))

    df = df.iloc[:, 0:6]
    df.columns = ["Time", "Open", "High", "Low", "Close", "Volume"]
    df.set_index("Time", inplace=True)
    df.index = pd.to_datetime(df.index, unit="ms")
    df = df.astype(float)
    return df
```

```python
interval = "5m"
symbol = "ETHUSDT"
begin_date = "2024-01-28T00:00:00Z"
end_date = "2024-01-29T00:10:00Z"

data = get_actual_data(symbol, time_frame=interval, begin_date = begin_date, end_date = end_date)
```
